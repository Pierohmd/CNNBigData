```python
matrice_path = 'data/MATRICE PENDOLARISMO 2011/matrix_pendo2011_10112014.txt'
comuni_path = 'data/MATRICE PENDOLARISMO 2011/elenco_codici/comuni_italiani_1_01_11.csv'
sovracomunali_path = 'data/MATRICE PENDOLARISMO 2011/elenco_codici/codici_sovracomunali.csv'
lines = []
with open(matrice_path, 'r') as f:
    for line in f:
        lines.append(line)
        
comuni_lines = []
with open(comuni_path, 'r') as f:
    for line in f:
        comuni_lines.append(line)

sovracomuni_lines = []
with open(sovracomunali_path, 'r', encoding='latin_1') as f:
    for line in f:
        sovracomuni_lines.append(line)
```

```python
regioni = {}
province = {}
comuni = {}
# comuni
for i, line in enumerate(comuni_lines):
    if i >= 5:
        splitted_line = line.split(',')
        comuni[splitted_line[3]] = splitted_line[7]
comuni['013249'] = 'Gravedona ed Uniti'
# regioni e province
for i, line in enumerate(sovracomuni_lines):
    if i >= 10:
        splitted_line = line.split(';')
        if not splitted_line[7] in regioni: regioni[splitted_line[5]] = splitted_line[7]
        if not splitted_line[12] in province: province[splitted_line[16][:-1]] = splitted_line[12]
```

```python
for line in lines:
    com_res = line[8:11] # comune di residenza
    prov_res = line[4:7] # provincia di residenza
    tipo = line[17:18] # 1 stesso comune, 2 altro comune ita, 3 estero
    com_dest =  line[23:26] # comune di destinazione (lavoro o studio)
    prov_dest = line[19:22] # provincia di destinazione (lavoro o studio)
    mezzo = line[31:33] # 01 treno, ....
    stima_flusso = line[38:38+12]
    flusso = line[50:50+10]
    if tipo == '2' and mezzo == '01':
        if comuni[prov_res+com_res]=='Roma':
            if comuni[prov_dest+com_dest]=='Frosinone':
                print(comuni[prov_res+com_res],comuni[prov_dest+com_dest],stima_flusso)
```
